# Imports & Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

# Make src importable
root = Path("..").resolve()
sys.path.append(str(root / "src"))

# Import functions
from functions import load_data, train_val_split, summarize_split
```

# Data Loading and Preprocessing (Task 1)

## Load Data

```python
dev_path = root / "data" / "development_data.csv"
eval_path = root / "data" / "evaluation_data.csv"

dev_df, eval_df = load_data(dev_path, eval_path)

dev_df.shape, eval_df.shape
```

## Train/Validation Split

```python
train_df, val_df = train_val_split(dev_df)
train_df.shape, val_df.shape
```

## Missing Values Overview

```python
missing_perc = dev_df.isna().mean().sort_values(ascending=False)
missing_perc
```

## Summary Statistics

```python
summarize_split(train_df, "Train")
summarize_split(val_df, "Validation")
summarize_split(eval_df, "Evaluation")
```

## Age Distribution (Development Set)

```python
plt.figure(figsize=(6,4))
sns.histplot(dev_df["age"], bins=30, kde=True)
plt.xlabel("Age (years)")
plt.ylabel("Count")
plt.title("Age Distribution – Development Set")
plt.tight_layout()
plt.show()
```
